```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame(columns = ['date_trunc','set_point_ventilation_unit', 'house no.'])
for i in range (1,11):
    path = '/Users/xinyuyangren/Desktop/Ventilation_Position/' + str(i) + '_ventpos.csv'
    ventpos = pd.read_csv(path)
    #ventpos.date_trunc = pd.to_datetime(ventpos.date_trunc)
    #ventpos.index = ventpos['date_trunc']
    #davg = ventpos.resample('D', how='mean')
    ventpos['house no.'] = i
    df = df.append(ventpos)
```

```python
df.date_trunc = pd.to_datetime(df.date_trunc)
t = pd.DatetimeIndex(df.date_trunc)
hr = t.hour
df['HourOfDay'] = hr
month = t.month
df['Month'] = month
year = t.year
df['Year'] = year
df_2013 = df[df['Year'] == 2013]
df_2014 = df[df['Year'] == 2014]
df1 = pd.concat([df_2013,df_2014])
#df1 = df1.dropna()
```
